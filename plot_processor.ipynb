```python
import matplotlib.pyplot as plt
from PIL import Image

# Reading the saved images
img1 = Image.open('doc/figures/2012_ERP_vs_Public_hospital_separations.png')
img2 = Image.open('doc/figures/Distance_to_nearest_public_hospital_with_ED_vs_Cat_4_5_ED_presentations.png')
img3 = Image.open('doc/figures/Population_density_vs_Distance_to_nearest_public_hospital.png')

# Getting the size of each plot
width1, height1 = img1.size
width2, height2 = img2.size
width3, height3 = img3.size

# Assuming all images have the same height, you can calculate the size of the final stitched image
stitched_width = width1 + width2 + width3  # Width of all three images in a row
stitched_height = max(height1, height2, height3)  # The height is the maximum of all three images

# Creating a new blank image for stitching
stitched_img = Image.new('RGB', (stitched_width, stitched_height))

# Pasting the three images in a single row
stitched_img.paste(img1, (0, 0))
stitched_img.paste(img2, (width1, 0))
stitched_img.paste(img3, (width1 + width2, 0))

# Saving and showing the stitched image
stitched_img.save('doc/figures/stitched_img_hospitals.png')
stitched_img.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the saved plot images
img1 = Image.open('doc/figures/public_hospitals_map.png')
img2 = Image.open('doc/figures/stitched_img_hospitals.png')

# Get the sizes of the images
width1, height1 = img1.size
width2, height2 = img2.size

# Create a new image that is tall enough to hold both images vertically
stitched_height = height1 + height2
stitched_width = max(width1, width2)

# Create a new blank image with the combined size
stitched_img = Image.new('RGB', (stitched_width, stitched_height))

# Paste the first image on the top
stitched_img.paste(img1, (3, 0))

# Paste the second image on the bottom
stitched_img.paste(img2, (0, height1))

# Display the combined image using matplotlib
plt.figure(figsize=(15, 15))  # Adjust the figsize as needed
plt.imshow(stitched_img)
plt.axis('off')  # Hide axis for a cleaner image display
plt.show()

# Optionally, save the stitched image
stitched_img.save('combined_vertical_plots.png')
stitched_img.show()
```
