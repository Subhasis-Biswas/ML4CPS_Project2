```python
# use geographical coordinates of the hospitals to calculate the co-ordinates relative to Melbourne GPO

# Melbourne GPO co-ordinates
melb_gpo = (-37.81384, 144.963028)

# compute (X,Y) co-ordinates of the hospitals relative to Melbourne GPO, with Melbourne GPO as the origin
dLat = public_hospital_locs['Latitude'] - melb_gpo[0]
dLon = public_hospital_locs['Longitude'] - melb_gpo[1]

# turn to cartesian co-ordinates
# use geopy

from geopy.distance import geodesic

def get_cartesian_coords(lat,lon):
	dN=geodesic((lat,lon),(melb_gpo[0],lon)).km
	dE=geodesic((lat,lon),(lat,melb_gpo[1])).km

	# adjust for the sign
	if lat < melb_gpo[0]:
		dN*=-1
	if lon < melb_gpo[1]:
		dE*=-1

	return dE,dN

public_hospital_locs['cart_x'],public_hospital_locs['cart_y']=zip(*public_hospital_locs.apply(lambda x: get_cartesian_coords(x['Latitude'],x['Longitude']),axis=1))

public_hospital_locs
```

## Question 1

```python
y=hospital_data['Public hospital separations, 2012-13'].values
X=data['2012 ERP, total'].values
X_flat=X
X=X.reshape(-1,1)
```

```python
np.random.seed(42)
# print correlation between X and y
correlation=np.corrcoef(X_flat,y)[0,1]
print(f'Correlation between X and y: {correlation}')

# fit a linear regression model
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X,y)

# print the r^2 score
r2=model.score(X,y)
print(f'R^2 score: {r2}')

# check autocorrelation of residuals
# compute residuals
y_pred=model.predict(X)
residuals=y-y_pred

# compute Moran's I
moran=Moran(residuals,w,permutations=9999)
print(f"Moran's I: {moran.I}")
print(f"P-value: {moran.p_sim}")



# do a scatter plot of X vs y
plt.figure(figsize=(5,5))
plt.scatter(X,y)
# plot the regression line
plt.plot(X,model.predict(X),color='black')
plt.xlabel('2012 ERP, total')
plt.ylabel('Public hospital separations, 2012-13')
plt.title('2012 ERP, total vs Public hospital separations, 2012-13',fontsize=10)
plt.legend(['Data','Regression Line'])

plt.text(0.65,0.01,f'R^2: {r2:.2f}',transform=plt.gca().transAxes)
plt.text(0.65,0.05,f'Correlation: {correlation:.2f}',transform=plt.gca().transAxes)
plt.text(0.65,0.10,f"Moran's I in Residual: {moran.I:.2f}",transform=plt.gca().transAxes)
plt.text(0.65,0.15,f"P-value: {moran.p_sim:.5f}",transform=plt.gca().transAxes)

plt.savefig('../doc/figures/2012_ERP_vs_Public_hospital_separations.png')

plt.show()
plt.clf()
```

```python
model = spreg.ML_Lag(y, X, w, name_y='Public hospital separations, 2012-13', name_x=['2012 ERP, total'], name_w='KNN', name_ds='Hospital Data',slx_lags=1)

print(model.summary)
```
